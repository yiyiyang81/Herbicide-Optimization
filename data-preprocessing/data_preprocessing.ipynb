```python
with open('corn.html', 'r', encoding='utf-8') as file:
    html_content = file.read()
from bs4 import BeautifulSoup

soup = BeautifulSoup(html_content, 'html.parser')
# Find all <th> tags
paragraphs = soup.find_all('th')

# Extract the text content from each <p> tag
target_labels = [chr(i) + '.' for i in range(ord('A'), ord('Z') + 1)]
target_labels += [chr(i) + chr(i)+ '.' for i in range(ord('A'), ord('H') + 1)]

current_label = None
next_label = None
name = []
quantity = []

import re
def contains_only_caps_or_asterisks(text):
    # Use a regular expression to match only capital letters (A-Z) or asterisks (*)
    return re.match(r'^[A-Z1-9*,-]*$', text) is not None or re.match(r'^[A-Z*]$', text) or "HALOSULFURON/DICAMBA*" in text

for p in paragraphs:
    if contains_only_caps_or_asterisks(p.text) and p.text:
        if p.text != "NA" and p.text!='12':
            name.append(p.text.replace('*',''))
            current_label = p.text
    elif current_label and re.search(r'\d',p.text):
        quantity.append(p.text)
        current_label = None
    elif current_label:
        quantity.append('N/A')
        current_label = None

types = ['before' for i in range(6)]
types += ['at' for i in range(6,13)]
types += ['after' for i in range(13,27)]

lower_bound = []
upper_bound = []

for item in quantity:
    if item == 'N/A':
        lower_bound.append('N/A')
        upper_bound.append('N/A')
    elif "â€“" in item:
        bounds = item.split("â€“")
        lower = float(bounds[0].strip())
        upper = float(bounds[1].split()[0].strip())
        lower_bound.append(lower)
        upper_bound.append(upper)
    else:
        bound = float(item.split()[0])
        lower_bound.append(bound)
        upper_bound.append(bound)


herbicidelist = pd.DataFrame({'name':name,'lower':lower_bound,'upper':upper_bound,'type':types})
herbicidelist.to_csv('corn_herbicide_restriction.csv',index=False)
```

```python
with open('cotton.html', 'r', encoding='utf-8') as file:
    html_content = file.read()
from bs4 import BeautifulSoup

soup = BeautifulSoup(html_content, 'html.parser')
# Find all <th> tags
paragraphs = soup.find_all('th')

# Extract the text content from each <p> tag
target_labels = [chr(i) + '.' for i in range(ord('A'), ord('Z') + 1)]
target_labels += [chr(i) + chr(i)+ '.' for i in range(ord('A'), ord('H') + 1)]

current_label = None
next_label = None
name = []
quantity = []

import re
def contains_only_caps_or_asterisks(text):
    # Use a regular expression to match only capital letters (A-Z) or asterisks (*)
    return re.match(r'^[A-Z1-9()\s*,-]*$', text) is not None or re.match(r'^[A-Z*]$', text) or "HALOSULFURON/DICAMBA*" in text

for p in paragraphs:
    # display(p.text.strip())
    
    string = p.text.strip()
    if contains_only_caps_or_asterisks(string) and string:
        # print(string)
        if string != "NA" and string !='12':
            name.append(string.replace('*',''))
            current_label = string
    elif current_label and re.search(r'\d',string):
        quantity.append(string)
        current_label = None
    elif current_label:
        quantity.append('N/A')
        current_label = None

time = ['before' for i in range(2)]
time += ['after' for i in range(2,6)]
time += ['N/A' for i in range(6,20)]
time += ['before' for i in range(20,24)]
time += ['after' for i in range(24,32)]
time += ['N/A' for i in range(32,36)]

types = ['winter_fallow_beds' for i in range(6)]
types += ['preplant' for i in range(6,10)]
types += ['postplant' for i in range(10,20)]
types += ['layby' for i in range(20,32)]
types += ['pre_harvest' for i in range(32,36)]

lower_bound = []
upper_bound = []

for item in quantity:
    if item == 'N/A':
        lower_bound.append('N/A')
        upper_bound.append('N/A')
    elif "â€“" in item:
        bounds = item.split("â€“")
        lower = float(bounds[0].strip())
        upper = float(bounds[1].split()[0].strip())
        lower_bound.append(lower)
        upper_bound.append(upper)
    elif "Up to" in item:
        lower_bound.append(0)
        upper_bound.append(float(item.split()[2]))
    else:
        # print(item)
        bound = float(item.split()[0])
        lower_bound.append(bound)
        upper_bound.append(bound)

herbicidelist = pd.DataFrame({'name':name,'lower':lower_bound,'upper':upper_bound, 'time':time, 'type':types
                              })
display(herbicidelist)
herbicidelist.to_csv('cotton_herbicide_restriction.csv',index=False)
```
